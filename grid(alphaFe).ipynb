```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
alpha0_table=np.loadtxt('alpha0')
```

```python
hb=alpha0_table[:,15]
Mgb=alpha0_table[:,19]
Fe5270=alpha0_table[:,20]
Fe5335=alpha0_table[:,21]
def MgFe(Mgb,  Fe5270, Fe5335):
    MgFe=(Mgb*(0.72*Fe5270+0.28*Fe5335))**0.5
    return MgFe
y=MgFe(Mgb, Fe5270, Fe5335)
#python shows sqrt(negative number)???
```

```python
plt.scatter(hb,y)
plt.xlabel('MgFe',fontsize=18)
plt.ylabel(r'$H_{\beta}$',fontsize=18)
plt.title(r'${\alpha}$ / Fe=0',fontsize=20)
ZH12=alpha0_table[:,1]
Year12=alpha0_table[:,0]
for value in np.unique(ZH12):
    mask=ZH12==value
    plt.plot(hb[mask],y[mask],color='blue')
for value in np.unique(Year12):
    mask=Year12==value
    plt.plot(hb[mask],y[mask],color='green')
```
