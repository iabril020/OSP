```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
%matplotlib inline
```

```python
#load data TMB MODEL
allyears_table=np.loadtxt('allyears')
```

```python
allyears_table.shape
```

```python
#select data of age=8
year8=[]
for i in range(allyears_table.shape[0]):
    if allyears_table[i,0]==8:
        year8.append(allyears_table[i,:])
year8=np.array(year8)
with open('year8','w') as  output:
    output.write(str(year8))
```

```python
#select data of age=12
year12=[]
for i in range(allyears_table.shape[0]):
    if allyears_table[i,0]==12:
        year12.append(allyears_table[i,:])
year12=np.array(year12)
with open('year12','w') as output:
    output.write(str(year12))
```

```python
#definition of new index
def MeanFe(Fe5270,Fe5335):
    MeanFe=(Fe5270+Fe5335)/2
    return MeanFe
```

```python
#extraction of the theoretical indexes from the model. calculation of indexes
Mgb_8=year8[:,19]
Fe5270_8=year8[:,20]
Fe5335_8=year8[:,21]
Mgb_12=year12[:,19]
Fe5270_12=year12[:,20]
Fe5335_12=year12[:,21]
x8=MeanFe(Fe5270_8,Fe5335_8)
x12=MeanFe(Fe5270_12,Fe5335_12)
y8=Mgb_8
y12=Mgb_12
```

```python
#[Z/H]-[alpha/Fe] grids of the TMB model at 8years
plt.scatter(x8,y8, color='mediumvioletred')
plt.xlabel('<Fe>',fontsize=18)
plt.ylabel('Mgb',fontsize=18)
plt.title('8Gyr',fontsize=20)

#ploting lines of constant metalicity and alpha/iron
ZH8=year8[:,1]
ZHpar=[]
AFe8=year8[:,2]
AFepar=[]
for value in np.unique(ZH8):
    mask=ZH8==value
    ZHpar.append(value)
    plt.plot(x8[mask],y8[mask],color='blue')
for value in np.unique(AFe8):
    mask=AFe8==value
    AFepar.append(value)
    plt.plot(x8[mask],y8[mask],color='green')

print(ZHpar)
print(AFepar)
#(HINDEX,MgINDEX,Fe5270INDEX,Fe5335INDEX)=(1.7108503714616412, 2.74202605338889, 2.0191738473455403, 1.4068659164397377)
#(Hb,MgFe,MeanFe)= 1.7108503714616412, 2.250892554511577, 1.713019881892639)

#final values from scale.py of VCC1407_M21
plt.plot(1.713019881892639, 2.74202605338889,'o')

plt.text(0.3,1.2, -2.25, fontsize=12)
plt.text(1.2,0.8,-1.35,fontsize=12)
plt.text(2.7,1.6,-0.33, fontsize=12)
plt.text(3.22,2.5,0.00,fontsize=12)
plt.text(4,3,0.35,fontsize=12)
plt.text(4.5,3.5,0.67,fontsize=12)
plt.text(4.5,4.5,-0.3,fontsize=12)
plt.text(4,5.2,0.0,fontsize=12)
plt.text(3.5,6,0.3,fontsize=12)
plt.text(3,6.2,0.5,fontsize=12)

green_patch=mpatches.Patch(color='green',label="alpha/Fe constant  ")
blue_patch=mpatches.Patch(color='blue',label="Z/H constant \dex")
dot=mlines.Line2D([],[],color='blue', marker='o',markersize=10,label='VCC1407')
plt.legend(handles=[green_patch,blue_patch,dot])

plt.savefig('8Gyr')
plt.show()
```
